# Anomaly Detection using Isolation Forest (scikit-learn)

Source: https://blog.paperspace.com/anomaly-detection-isolation-forest/  
Update: https://github.com/jehopark/anomaly_detection.git

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
url = "https://raw.githubusercontent.com/jehopark/anomaly_detection/main/Salaries.csv" # data with outliers
url2 = "https://raw.githubusercontent.com/jehopark/anomaly_detection/main/Salaries_na.csv" # data with missing values
df = pd.read_csv(url)
df.head(11)
```

```python
# model here
model=IsolationForest(n_estimators=50, max_samples='auto', contamination=float(0.03),max_features=1.0)
model.fit(df[['salary']])
```

```python
df['scores']=model.decision_function(df[['salary']])
df['anomaly']=model.predict(df[['salary']])
df.head(20)
```

```python
# print anomaly
anomaly=df.loc[df['anomaly']==-1]
anomaly_index=list(anomaly.index)
print(anomaly)
```
